### Bài 9:
 Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay 
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 

```python
n = int(input())
for i in range (1, n + 1):
    if (n**0.5)%i == 0:
        print("La so chinh phuong")
        break
    else:
        print("Khong la so chinh phuong")
        break
```

### Bài 10: 
Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. 
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…

```python
a = int(input())
sum = 0
for i in range(1, (a//2)+1):
    if a%i==0:
        sum+=i
if sum <= a:
    print(a, "La so hoan hao")
else:
    print(a, "Khong la so hoan hao")
```

### Bài 14: 
Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 


```python
x = int(input())
if x == -1:
    print("Dung nhap khi nhap -1")
else:
    max_lon = x
    min_nho = x
    while True:
        x = int(input())
        if x == -1:
            break
        if x > max_lon:
            max_lon = x
        if x < min_nho:
            min_nho = x
print("So lon nhat", max_lon)
print("So nho nhat", min_nho)
```

### Bài 16:
 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu 
chữ số lẻ 

```python
n = int(input())
chan = 0
le = 0
while n>0:
    digit = n%10
    n//=10
    if digit%2 == 0:
        chan+=1
    else:
        le+=1
print("so chu so chan", chan)
print("so chu so le", le)
```

```python
n = input()
chan = 0
le = 0
for x in n:
    if int(x)%2 == 0:
        chan+=1
    else:
        le+=1
print("so chu so chan", chan)
print("so chu so le", le)
```

### Bài 19: 
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input())
a = 1
b = 1
while True:
    c = a+b
    if c>A:
        break
    a = b 
    b = c
print(b)
```
